# 02 · Feature Engineering

**ADS 504 — Bitcoin Price Direction**

Build the technical indicators and construct the classification target.

> ⚠️ Every feature must use only information available at or before the current
> candle. The target looks one step into the future — do not leak it.

## 1. Setup & load cleaned data

## 2. Target variable

`direction = 1 if next_close > current_close else 0` (shift close by -1).
Report the class balance.

## 3. Trend features

Simple / exponential moving averages (SMA, EMA) and crossovers.

## 4. Momentum features

RSI, MACD (line / signal / histogram).

## 5. Volatility features

Bollinger Bands (upper / lower / width), rolling standard deviation.

## 6. Returns & volume features

Log returns, lagged returns, volume change.

## 7. Assemble & rank features

Drop warm-up NaNs from the indicators, review correlations, and rank/select
candidate features. Save the modeling table for `03_modeling`.
